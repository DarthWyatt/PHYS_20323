```python
import numpy as np
from numpy import arange
from pylab import plot,xlabel,ylabel,show
from matplotlib import pyplot as plt
def run_simulation():

    NAt = 25000              # Constants
    NBi215 = 0
    NBi211 = 0
    NRn = 0                    
    NPb211 = 0
    NPb207 = 0
    NTl = 0

    tau_At = 6*60            # Half life of elements in seconds
    tau_Bi215 = 7.6*60
    tau_Bi211 = 2.14*60
    tau_Rn = 140
    tau_Pb211 = 36.1*60
    tau_Tl = 4.77*60

    h = 1.0                    # Size of time-step in seconds

    p_At = 1 - 2**(-h/tau_At)  # Probability of decay in one step
    p_Bi215 = 1 - 2**(-h/tau_Bi215)
    p_Bi211 = 1 - 2**(-h/tau_Bi211)
    p_Rn = 1 - 2**(-h/tau_Rn)
    p_Pb211 = 1 - 2**(-h/tau_Pb211)
    p_Tl = 1 - 2**(-h/tau_Tl)

    tmax = 200                # Total time

    red_count = 0
    blue_count = 0
    green_count = 0
    black_count = 0
    red_energy = 0.0
    blue_energy = 0.0
    green_energy = 0.0
    black_energy = 0.0

    tpoints = arange(0.0,tmax,h)
    Atpoints = [] # empty array to strore # of atoms at each time step
    Bi215points = [] 
    Bi211points = []
    Rnpoints = []
    Pb211points = []
    Pb207points = []
    Tlpoints = []

    # Main loop
    for t in tpoints:    
        Atpoints.append(NAt) # append starting # of atoms
        Bi215points.append(NBi215) 
        Bi211points.append(NBi211)
        Rnpoints.append(NRn)
        Pb211points.append(NPb211)
        Pb207points.append(NPb207)
        Tlpoints.append(NTl)

        # Calculate the number of atoms that decay
        decay_Bi211 = 0
        for h in range(NBi211):
            if np.random.random() < p_Bi211:
                if np.random.random() < .13:
                    NPb207 += 1
                    NBi211 -= 1
                    black_count += 1
                    black_energy += 3.0
                else:
                    NTl += 1
                    NBi211 -= 1
                    red_count += 1
                    red_energy += 5.0
    
        decay_Tl = 0
        for h in range(NTl):
            if np.random.random() < p_Tl:
                decay_Tl += 1
                blue_count += 1
                blue_energy += 0.5
        NTl -= decay_Tl
        NPb207 += decay_Tl

        decay_Pb211 = 0
        for h in range(NPb211):
            if np.random.random() < p_Pb211:
                decay_Pb211 += 1
                blue_count += 1
                blue_energy += 0.5
        NPb211 -= decay_Pb211
        NBi211 += decay_Pb211

        decay_Rn = 0
        for h in range(NRn):
            if np.random.random() < p_Rn:
                decay_Rn += 1
                black_count += 1
                black_energy += 3.0
        NRn -= decay_Rn
        NPb211 += decay_Rn

        decay_Bi215 = 0
        for h in range(NBi215):
            if np.random.random() < p_Bi215:
                decay_Bi215 += 1
                green_count += 1
                green_energy += 12.0
        NBi215 -= decay_Bi215
        NPb211 += decay_Bi215

        decay_At = 0
        for i in range(NAt):
            if np.random.random() < p_At:
                if np.random.random() < .90:
                    NBi215 += 1
                    NAt -= 1
                    red_count += 1
                    red_energy += 5.0
                else:
                    NRn += 1
                    NAt -= 1
                    blue_count += 1
                    blue_energy += 0.5

    # Make the graph
    fig = plt.figure(figsize=(12,7))
    ax = fig.add_subplot(111)
    ax.plot(tpoints,Atpoints, c='m') 
    ax.plot(tpoints,Bi215points, c='r') 
    ax.plot(tpoints,Bi211points, c='g')
    ax.plot(tpoints,Rnpoints, c='b')
    ax.plot(tpoints,Pb211points, c='c')
    ax.plot(tpoints,Pb207points, c='y')
    ax.plot(tpoints,Tlpoints, c='k')
    # add the labels
    ax.set_xlabel("Time (s)",size=20)   # allows LaTeX style formating
    ax.set_ylabel("Number of atoms",size=20)   # allows LaTeX style formating

    ax.xaxis.set_tick_params(labelsize=20)
    ax.yaxis.set_tick_params(labelsize=20)
    plt.show()

    print("Radiation Totals:")
    print(f"Alpha decay: {red_count}, Energy = {red_energy:.2f} MeV")
    print(f"Beta decay: {blue_count}, Energy = {blue_energy:.2f} MeV")
    print(f"Z decay: {green_count}, Energy = {green_energy:.2f} MeV")
    print(f"R decay: {black_count}, Energy = {black_energy:.2f} MeV")

    total_energy = red_energy + blue_energy + green_energy + black_energy
    print(f"\nTotal Radiation Energy = {total_energy:.2f} MeV\n")
    return red_energy, blue_energy, black_energy, green_energy, total_energy

num_runs = 10
results = []

for i in range(num_runs):
    results.append(run_simulation())

results = np.array(results)

red_vals = results[:, 0]
blue_vals = results[:, 1]
black_vals = results[:, 2]
green_vals = results[:, 3]
total_vals = results[:, 4]

print("\nStats Across 10 Runs:")
print(f"Alpha (red, 5 MeV): Average = {red_vals.mean():.2f}, Standard Deviation = {red_vals.std():.2f}")
print(f"Beta (blue, 0.5 MeV): Average = {blue_vals.mean():.2f}, Standard Deviation = {blue_vals.std():.2f}")
print(f"R (black, 3 MeV): Average = {black_vals.mean():.2f}, Standard Deviation = {black_vals.std():.2f}")
print(f"Z (green, 12 MeV): Average = {green_vals.mean():.2f}, Standard Deviation = {green_vals.std():.2f}")
print(f"\nTotal Radiation Energy: Average = {total_vals.mean():.2f}, Standard Deviation = {total_vals.std():.2f}")


stats = {
    "alpha": (red_vals.mean(), red_vals.std()),
    "beta": (blue_vals.mean(), blue_vals.std()),
    "R": (black_vals.mean(), black_vals.std()),
    "Z": (green_vals.mean(), green_vals.std()),
}
materials = {
    "alpha": {"Material": "Wood", "MeV_per_cm": 7000, "Cost": 1.00},
    "beta": {"Material": "Water", "MeV_per_cm": 2000, "Cost": 2.00},
    "R": {"Material": "Gold", "MeV_per_cm": 6000, "Cost": 200.00},
    "Z": {"Material": "Lead", "MeV_per_cm": 3000, "Cost": 20.00},
}
print("\nShielding Requirements (Mean + 3):")
total_cost = 0

for rad_type in ["alpha", "beta", "R", "Z"]:
    mean, std = stats[rad_type]
    design_energy = mean + 3*std
    m = materials[rad_type]
    thickness = design_energy/m["MeV_per_cm"]
    cost = thickness*m['Cost']
    total_cost += cost

    print(f"\n{rad_type.upper()} Radiation:")
    print(f"Material: {m['Material']}")
    print(f"Design Energy: {design_energy:.2f} MeV")
    print(f"Required Thickness: {thickness:.4f} cm")
    print(f"Cost: ${cost:.2f}")

print(f"\nTotal Shielding Cost: ${total_cost:.2f}")
```
